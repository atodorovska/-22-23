The first step is detecting the count and % of missing values in every dataset feature, so that we have an idea about the distribution of missing values.

```python
train_df.isnull()
```

```python
train_missing_values_count = train_df.isnull().sum()
train_missing_values_count
```

```python
train_missing_values_percentage = 100 * train_missing_values_count / len(train_df)
train_missing_values_percentage
```

```python
train_missing = pd.concat([train_missing_values_count, train_missing_values_percentage], axis=1)
train_missing.columns = ['# of missing values', '% of missing values']
train_missing
```

```python
test_missing_values_count = test_df.isnull().sum()
test_missing_values_count
```

```python
test_missing_values_percentage = 100 * test_missing_values_count / len(test_df)
test_missing_values_percentage
```

```python
test_missing = pd.concat([test_missing_values_count, test_missing_values_percentage], axis=1)
test_missing.columns = ['# of missing values', '% of missing values']
test_missing
```

Dropping the missing values

```python
train_df_copy = train_df.copy()
```

```python
#Note: Pandas will skip the missing values to calculate the mean.
train_df_copy['Age'].mean()
```

```python
train_df_copy.dropna(subset=['Age'], how='any', inplace=True)
```

Imputing the missing values

```python
from sklearn.impute import SimpleImputer
```

```python
train_df_copy = train_df.copy()
```

```python
imputer = SimpleImputer(strategy='constant', fill_value=0)
```

```python
#Note: fit -> means look at the null values and transform -> means to replace with 0s in this case.
train_df_copy.iloc[:,:] = imputer.fit_transform(train_df_copy)
```

```python
train_df_copy.isnull().sum()
```

```python
train_df_copy = train_df.copy()
```

```python
imputer = SimpleImputer(strategy='most_frequent')
```

```python
train_df_copy.iloc[:,:] = imputer.fit_transform(train_df_copy)
```
